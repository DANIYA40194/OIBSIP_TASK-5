```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Load the dataset
data = pd.read_csv('/content/sales data file.csv')

# Explore the dataset to understand its structure and missing values
print(data.head())
print(data.info())
```

```python
# Check for missing values
print(data.isnull().sum())

# If there are missing values, handle them appropriately, e.g., drop rows with missing values
data.dropna(inplace=True)
```

```python
# Set the style and palette
sns.set(style="ticks", palette="pastel")

# Create pairplot with custom color palette
sns.pairplot(data, diag_kind='kde', plot_kws={'alpha': 0.6})

# Show the plot
plt.show()
```
